### Q1. What is Bayes' theorem?

Ans:In statistics and probability theory, the Bayes’ theorem (also known as the Bayes’ rule) is a mathematical formula used to determine the conditional probability of events. Essentially, the Bayes’ theorem describes the probability of an event based on prior knowledge of the conditions that might be relevant to the event.

### Q4. What is the relationship between Bayes' theorem and conditional probability?

Ans:Bayes' theorem relates conditional probabilities P(A∣B) and P(B∣A) to each other, along with the prior probabilities of events A and B. It provides a framework to calculate the probability of one event given the probability of another event occurring, along with their conditional probabilities.

### Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?

Ans:The choice of Naive Bayes classifier depends on the nature of the problem and the assumptions about the independence of features:

* Gaussian Naive Bayes: Suitable for continuous features that follow a Gaussian distribution.

* Multinomial Naive Bayes: Typically used for text classification tasks with discrete features (e.g., word counts).

* Bernoulli Naive Bayes: Suitable for binary feature vectors, often used in document classification tasks where the presence or absence of features is considered.
