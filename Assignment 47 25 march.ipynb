### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

df['Size'] = df['Size'].str.replace('M', '').str.replace('k', '')
df['Size'] = df['Size'].str.replace(',', '')
df['Size'] = df['Size'].replace('Varies with device', None)
df['Size'] = df['Size'].replace('1000+', '1000').astype(float)

df_filtered = df.dropna(subset=['Size', 'Rating'])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Size', y='Rating',hue='Rating')

plt.title('App Rating vs. Size')
plt.xlabel('App Size')
plt.ylabel('App Rating')

plt.show()
```

### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

df['Price'] = df['Price'].replace('Everyone', np.nan)

df['Price'] = df['Price'].str.replace('$', '').astype(float)

avg_prices = df.groupby('Type')['Price'].mean()

plt.figure(figsize=(8, 6))
plt.bar(avg_prices.index, avg_prices)

plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')

plt.show()
```

### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

df_sorted = df.sort_values('Installs', ascending=False)

top_10_apps = df_sorted.head(10)

app_frequency = top_10_apps['App'].value_counts().reset_index()
app_frequency.columns = ['App', 'Frequency']

print(app_frequency)
```

### Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
```

```python
print(df['Category'].unique())
```

```python
df = df.dropna()

df = df.drop_duplicates()

df['Installs'] = df['Installs'].apply(lambda x: int(x.replace('+', '').replace(',', '')))
df['Reviews'] = df['Reviews'].apply(int)
```

```python
category_installs = df.groupby('Category')['Installs'].sum()

sorted_categories = category_installs.sort_values(ascending=False)

print(sorted_categories.head(10))
```

```python
top_categories = sorted_categories.head(3)

recommended_categories_df = pd.DataFrame({'Category': top_categories.index, 'Installs': top_categories.values})

print("Recommended App Categories:")
print(recommended_categories_df)
```

### Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
```

```python
df = df.dropna()

df['Installs'] = df['Installs'].apply(lambda x: int(x.replace('+', '').replace(',', '')))
df['Reviews'] = df['Reviews'].apply(int)
```

```python
App_installs = df.groupby('App')['Installs'].sum()

App_ratings = df.groupby('App')['Rating'].mean()

App_reviews = df.groupby('App')['Reviews'].sum()

App_updates = df.groupby('App')['Last Updated'].count()

App_categories = df.groupby('App')['Category'].nunique()
```

```python
# Create a DataFrame with the analyzed metrics
Apps_df = pd.DataFrame({
    'Installs': App_installs,
    'Average Rating': App_ratings,
    'Reviews': App_reviews,
    'App Updates': App_updates,
    'App Categories': App_categories
})
```

```python
Apps_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

Apps_df['Installs'].nlargest(10).plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Apps by Total Installs')
plt.xlabel('App')
plt.ylabel('Total Installs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Installs vs. Average Rating
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Installs', y='Average Rating', data=Apps_df)
plt.title('App Performance: Installs vs. Average Rating')
plt.xlabel('Installs')
plt.ylabel('Average Rating')
plt.show()
```
