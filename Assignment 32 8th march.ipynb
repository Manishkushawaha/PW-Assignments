Q1. What is the Probability density function?

Ans:A Probability Density Function (PDF) is a mathematical function that describes the relative likelihood of observing a random variable (such as a continuous random variable) within a given range of values. The PDF is often denoted by the letter "f(x)" and is defined as follows:

For any continuous random variable X, the PDF f(x) is a function such that for any two values a and b, with a < b:

The probability that X takes on a value between a and b is given by the integral of f(x) over the interval [a, b], i.e., P(a <= X <= b) = ∫[a, b] f(x) dx.


The area under the curve of f(x) over the entire range of possible values of X is equal to 1, i.e., ∫(-∞, +∞) f(x) dx = 1.


Q2. What are the types of Probability distribution?

Ans:There are many types of Probability distributions, but some of the most common ones are:

Normal distribution: A continuous Probability distribution that is symmetrical around the mean and has a bell-shaped curve. Many natural phenomena follow this distribution, such as heights, weights, and IQ scores.

Binomial distribution: A discrete Probability distribution that describes the number of successes in a fixed number of independent trials with two possible outcomes. For example, the number of heads in 10 coin flips.

Poisson distribution: A discrete Probability distribution that describes the number of rare events occurring in a fixed interval of time or space. For example, the number of earthquakes per year in a certain region.

Exponential distribution: A continuous Probability distribution that describes the time between rare events occurring in a Poisson process. For example, the time between car accidents on a specific road.

Uniform distribution: A continuous Probability distribution that has a constant probability density function over a specific interval. For example, the distribution of a fair roll of a die.

Chi-squared distribution: A continuous Probability distribution that arises in hypothesis testing, often used to test whether two samples come from the same population.

Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import scipy.stats as stats
import seaborn as sns

def normal_pdf(x,mean,std_dev):
    
    pdf = stats.norm.pdf(x,mean,std_dev)
    
    return pdf

normal_pdf=normal_pdf(2, 0.5, 1.5)

print("the probability density function of a normal distribution is",normal_pdf)
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

def normal_pdf(x, mu, sigma):
   
    return stats.norm.pdf(x, mu, sigma)

def plot_normal_pdf(mu, sigma):
   
    x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
    y = normal_pdf(x, mu, sigma)
    plt.plot(x, y)
    plt.title("Normal Distribution (mean={}, std={})".format(mu, sigma))
    plt.xlabel("x")
    plt.ylabel("Probability density")
    plt.show()
plot_normal_pdf(2, 0.5)
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Ans:The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, each with the same probability of success. Here are some of its properties:

The distribution is discrete, meaning that the random variable can only take on integer values.

The distribution is characterized by two parameters: the number of trials n and the probability of success p.

The probability mass function of the binomial distribution is given by P(X = k) = (n choose k) * p^k * (1 - p)^(n-k), where X is the random variable representing the number of successes in n trials.

The mean of the distribution is mu = np, where n is the number of trials and p is the probability of success.

The variance of the distribution is sigma^2 = np(1-p).

Here are two examples of events where binomial distribution can be applied:

Flipping a coin: Suppose we flip a coin n times, and we want to know the probability of getting k heads. This is an example of a binomial distribution, where the probability of success p is 0.5 (assuming a fair coin), and the number of trials n is the number of flips.

Drug testing: Suppose a new drug is tested on n patients, and the probability of success is p. We want to know the probability that exactly k patients will respond positively to the drug. This is an example of a binomial distribution, where the number of trials n is the number of patients tested, and the probability of success p is the probability that a patient responds positively to the drug.

Q7. How Binomial distribution different from Poisson distribution?

Ans:The Binomial and Poisson distributions are both used to model the occurrence of events. However, they have different characteristics and are used in different situations.

Here are some key differences between the Binomial and Poisson distributions:

Number of trials: The Binomial distribution models the number of successes in a fixed number of trials, while the Poisson distribution models the number of events in a fixed amount of time or space.

Probability of success: In the Binomial distribution, the probability of success is constant across all trials, while in the Poisson distribution, the probability of an event occurring is assumed to be small and constant across all time or space intervals.

Relationship between mean and variance: In the Binomial distribution, the variance is equal to the mean times the complement of the probability of success, while in the Poisson distribution, the variance is equal to the mean.

Assumptions: The Binomial distribution assumes that the trials are independent and identically distributed, while the Poisson distribution assumes that events occur independently of one another and at a constant rate.

Applications: The Binomial distribution is commonly used to model the number of successes or failures in a fixed number of trials, such as the number of heads in a series of coin tosses. The Poisson distribution is commonly used to model the number of rare events occurring in a fixed amount of time or space, such as the number of customers arriving at a store in a given hour.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

Ans: In a normal distribution, the least frequent data appear in the tails of the distribution, farthest away from the mean. This is because the normal distribution is a symmetric distribution, with the mean located at the center of the distribution and the tails extending infinitely in both directions.

The normal distribution is characterized by two parameters, the mean (μ) and the standard deviation (σ). About 68% of the data in a normal distribution lie within one standard deviation of the mean, about 95% of the data lie within two standard deviations of the mean, and about 99.7% of the data lie within three standard deviations of the mean.

Therefore, the least frequent data in a normal distribution appear beyond three standard deviations from the mean, in the tails of the distribution. This is where the data are most spread out and the probability of observing a value in this region is relatively low.
